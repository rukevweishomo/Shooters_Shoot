# Test Scrapers
I wanted to start this project with some tests, to fully understand the scope of the API I was working with. My first instinct was to see how many players were in the database, and also to check for duplicates. I also did some tweaking to make sure the user can access a team's record and performance in a given season, or in the playoffs idf they wish. Finally, I wanted to explore the extent of the 'shotchartdetail' module by adding everything I've experimented with combined with the attributes of the module to give a conclusive overview of shots taken in basketball games across the NBA.

```python
#importing libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import time
from nba_api.stats.static import players
from nba_api.stats.static import teams
from nba_api.stats.endpoints import playercareerstats
pd.set_option('display.max_columns', 500)
pd.set_option('display.max_rows', 500)

player_input = input("Enter the name of the player: ")
# Fetching player data using nba_api
player_dict = players.get_players()
# Finding the player(s) based on the input name
matching_players = [
    player for player in player_dict if player["full_name"] == player_input
]
if not matching_players:
    raise ValueError(f"No player found with name {player_input}")

print(f"Found {len(matching_players)} player(s) matching '{player_input}':")
for idx, player in enumerate(matching_players, start=1):
    print(f"{idx}: {player['full_name']} (ID: {player['id']})")

if len(matching_players) == 1:
    player_output = matching_players[0]
else:
    selection = int(input("Multiple players found. Select one by number: "))
    player_output = matching_players[selection - 1]

career = playercareerstats.PlayerCareerStats(player_id=player_output["id"])
career.get_data_frames()[0]

```

```python
from nba_api.stats.static import teams
from nba_api.stats.endpoints import leaguegamefinder
from nba_api.stats.endpoints import teamgamelog
from nba_api.stats.library.parameters import Season, SeasonType

teams_dict = teams.get_teams()
team_input = input("Enter the name of the team (its abbreviation): ")
matching_team = [
    team for team in teams_dict if team["abbreviation"] == team_input
]
if not matching_team:
    raise ValueError(f"No team found with abbreviation {team_input}")
team_output = matching_team[0]

year = int(input("Enter the season year (e.g., 2022 for the 2022-23 season): "))
period = input("Enter the season type (e.g., Regular Season, Playoffs): ")

def format_year(year: int) -> str:
    return f"{year}-{str(year + 1)[-2:]}"

gamefinder = teamgamelog.TeamGameLog(team_id=team_output["id"], season=format_year(year), season_type_all_star=period)
gamefinder.get_data_frames()[0]
```

```python
from nba_api.stats.endpoints import shotchartdetail

player_input = input("Enter the name of the player (Type in '0' if you want information on all players): ")
# Fetching player data using nba_api
player_dict = players.get_players()
# Finding the player(s) based on the input name
if player_input == "0":
    player_id = 0
else:
    matching_players = [
        player for player in player_dict if player["full_name"] == player_input
    ]
    if not matching_players:
        raise ValueError(f"No player found with name {player_input}") 
    
    print(f"Found {len(matching_players)} player(s) matching '{player_input}':")
    for idx, player in enumerate(matching_players, start=1):
        print(f"{idx}: {player['full_name']} (ID: {player['id']})")
    if len(matching_players) == 1:
        player_output = matching_players[0]
        player_id = player_output["id"]
    else:
        selection = int(input("Multiple players found. Select one by number: "))
        player_output = matching_players[selection - 1]

team_input = input("Enter the abbreviated name of the team you want to analyze (Type in 0 if you want info on all teams): ")
if team_input == "0":
    team_id = 0
else:
    matching_team = [
        team for team in teams_dict if team["abbreviation"] == team_input
    ]
    if not matching_team:
        raise ValueError(f"No team found with abbreviation {team_input}")
    team_output = matching_team[0]
    team_id = team_output["id"]

year = int(input("Enter the season year (e.g., 2022 for the 2022-23 season): "))
period = input("Enter the season type (e.g., Regular Season, Playoffs): ")
context = input("Enter the context of the shot data you want to analyze (e.g., FGA, 3PA, etc): ")

def format_year(year: int) -> str:
    return f"{year}-{str(year + 1)[-2:]}"

shots = shotchartdetail.ShotChartDetail(
    team_id=team_id,
    player_id=player_id,
    season_nullable=format_year(year),
    season_type_all_star=period,
    context_measure_simple=context
)

df = shots.get_data_frames()[0]
df.head(10)
```
